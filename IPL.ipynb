```python
outliers=[]
def outlierFinder(data):
    threshold=3
    mean=np.mean(data)
    sd=np.std(data)
    for i in data:
        z_score = (i-mean)/sd
        if(z_score>threshold):
            outliers.append(i)
            
    return outliers
```

```python
outliers=outlierFinder(totalCol)
```
